## Load "session"


```python
import requests
import pandas as pd

# Fetch all sessions
url = "https://api.openf1.org/v1/sessions"
response = requests.get(url)
data = response.json()

# Convert to DataFrame
df = pd.DataFrame(data)

# Filter for race sessions in Spielberg
race_sessions_spielberg = df[(df['session_type'] == 'Race') & (df['location'] == 'Spielberg')]

# Display result
display(race_sessions_spielberg)
```

```python
for col in race_sessions_spielberg.columns:
    print(col)
```

```python
# For most drivers, lap_number = 1 is not a timed lap yet,
# but rather the out-lap or formation/start lap:

# Sometimes there is no official timing start for the first lap.
# if the car was stationary in the pit or just left the pit lane
```

```python
display(laps_driver_63)
```

```python
import requests
import pandas as pd

# ✅ Add both filters: session_key and driver_number
url = "https://api.openf1.org/v1/laps?session_key=9550&driver_number=1"

response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)

laps_driver_1 = df
```

```python
print(f"Num of Laps: {len(laps_driver_1)}")
```

```python
# lap_number means: "the car’s own count of crossing the timing line"
#
# Example:
# Driver   lap_number   date_start
#   1           2       2024-06-30T13:04:14.597Z
#  63           2       2024-06-30T13:04:16.930Z
#
# Here, lap 2 refers to the first flying lap after the race start.
# Because drivers have different grid positions, they cross the start line
# at slightly different times. So the same lap number can have different start times.
# This means Driver 1 crossed the line about 2.3 seconds before Driver 63.
```

```python
display(laps_driver_1)
```

```python
import requests
import pandas as pd

# ✅ Add both filters: session_key and driver_number
url = "https://api.openf1.org/v1/intervals?session_key=9550&driver_number=1"

response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)

intervals_driver_1 = df
```

```python
print(f"Num of Intervals: {len(intervals_driver_1)}")
```

```python
# In the OpenF1 intervals endpoint, an "interval" usually means:
# "How far behind this driver is compared to the car ahead or the race leader."
#
# Therefore:
# - The race leader has no car ahead → so there is no interval to report 
#   (or only a self-reference, which may not be stored).
# - All other drivers have intervals to the car in front or to the leader,
#   so they have many more interval records.
```

```python
display(intervals_driver_1)
```

## Load "position"

```python
import requests
import pandas as pd

url = "https://api.openf1.org/v1/position?session_key=9550"
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)

position = df

# Convert to datetime
position['date'] = pd.to_datetime(position['date'])

# Overwrite the specific datetime
position.loc[
    position['date'] == pd.Timestamp('2024-06-30 12:07:38.634000+00:00'),
    'date'
] = pd.Timestamp('2024-06-30 13:00:00.000000+00:00')

# Extract only the time part if needed
position['timestamp'] = position['date'].dt.time
```

```python
display(position)
```

```python
print(f"Num of position: {len(position)}")
```
